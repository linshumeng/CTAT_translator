```python
import pandas as pd
import re
import translators.server as tss
TARGET_LANG = 'es'
```

```python
# read the latest mass production table
# path_new =  r".\Output_cleaned_folder\6_5_cleaned.txt"
path_new = r".\Output_cleaned_folder\7-17_finalMassProduction_new_cleaned.txt"
table_new = pd.read_csv(path_new, sep="\t", index_col=0, keep_default_na=False)
```

```python
# replace variable with value in the latest mass production table
# iterate column and row in table_new
for column in range(table_new.shape[1]):
    for row in range(table_new.shape[0]):
        content_new = str(table_new.iloc[row, column])
        start_pattern = "%\("
        # count the number of the replacement in one variable(content_new)
        count = [match.start() for match in re.finditer(start_pattern, str(content_new))]
        for i in range(len(count)):
            # find the variable
            start_pattern = "%\("
            start_index = [match.start() for match in re.finditer(start_pattern, str(content_new))]
            end_pattern = "\)%"
            end_index = [match.start() for match in re.finditer(end_pattern, str(content_new))]
            # no need for replacement
            if start_index == []:
                continue
            else:
                variable = content_new[start_index[0]: end_index[0]+2]
                # find the corresponding column name, and then find the value
                column_name = table_new.columns[column]
                try:
                    value = table_new.loc[variable, column_name] 
                    # print(variable, value)
                    content_new = content_new.replace(variable, value)
                    table_new.iloc[row, column] = content_new
                except:
                    print(variable + " doesn't exist")
table_new.head()
```

```python
# Change column name
table_new_col_list = []
for i in table_new:
    table_new_col_list.append(i)
    table_new_col_list.append(i+"ESP")
len(table_new_col_list)
```

```python
# read the Greg's mass production table
# path_old = './Greg_table_folder/6.05 - Sheet1.csv'
path_old = './Greg_table_folder/7.17 - Sheet1.csv'
table_old = pd.read_csv(path_old, header = None)
# table_old.head()
# skip some rows, because they are translation for HTML elements
header_index = table_old.index[table_old[0] == 'Problem Name'].to_list()
header_index
# reload the csv
table_old = pd.read_csv(path_old, header = header_index)
table_old.columns
```

```python
# table_translated = pd.DataFrame(columns=table_old.columns, index=range(table_new.shape[0]))
# Create a new table, with the column from the Greg's mass production table and the index from the latest mass production table
table_translated = pd.DataFrame(columns=table_new_col_list, index=table_new.index)
table_translated
```

```python
# find the translation of the latest mass production table from the Greg's mass production table 
google_dict = {}
# for column in range(table_new.shape[1]):
for column in range(1):
    for row in range(table_new.shape[0]):
        content_new = str(table_new.iloc[row, column])
        column_name = table_new.columns[column]
        column_num = table_translated.columns.get_loc(column_name)
        # write the english column
        table_translated[column_name].iloc[row] = content_new
        if table_old.columns[(table_old == content_new).any()].empty:
            if content_new.isdigit() or content_new == '':
                print("digital or empty")
                content_translated = content_new
            elif content_new in google_dict:
                print("find translation in dict")
                content_translated = google_dict[content_new]
            else:
                print("use google translation")
                try:
                    translation = tss.google(content_new, from_language='en', to_language=TARGET_LANG)
                except:
                    translation = 'error'
                content_translated = '[google]' + translation
                google_dict[content_new] = content_translated
        # find translation in old table
        else:
            print("find translation in sheet")
            column_name_old = table_old.columns[(table_old == content_new).any()][0]
            column_num_old = table_old.columns.get_loc(column_name_old)
            content_translated = table_old[table_old[column_name_old] == content_new].iloc[0, column_num_old+1]
        # # print(content_translated)
        print(content_new, content_translated)
        table_translated.iloc[row, column_num+1] = content_translated
        # # content_translated
table_translated
```
